```python
search = DDGS()
notes = []

def save_note(text: str):
    notes.append(text)
    return "Заметка сохранена"

def show_notes(_):
    if len(notes) == 0:
        return "Заметок нет"
    return "\n".join(notes)

def calculator(expr: str):
    try:
        return str(eval(expr))
    except Exception as e:
        return f"Ошибка вычисления: {str(e)}"
```

```python
def web_search(query: str):
    try:
        results = []
        for r in search.text(query, max_results=3):
            results.append(f"{r['title']} — {r['href']}")
        return "\n".join(results) if results else "Ничего не найдено"
    except Exception as e:
        return f"Ошибка поиска: {str(e)}"
```

```python
tools = [
    Tool(
        name="WebSearch",
        func=web_search,
        description="Используй для поиска информации в интернете. Входные данные: поисковый запрос"
    ),
    Tool(
        name="Calculator",
        func=calculator,
        description="Используй для математических вычислений. Входные данные: математическое выражение"
    ),
    Tool(
        name="SaveNote",
        func=save_note,
        description="Используй для сохранения заметки. Входные данные: текст заметки"
    ),
    Tool(
        name="ShowNotes",
        func=show_notes,
        description="Используй для показа всех сохраненных заметок. Входные данные: любой текст"
    )
]
```

```python
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)
```

```python
def github_api(state):
    repo = state["repo"]
    url = f"https://api.github.com/repos/{repo}"

    try:
        r = requests.get(url)
        if r.status_code != 200:
            return {"error": "Repository not found"}

        data = r.json()
        return {
            "repo": repo,
            "description": data.get("description", "Описание отсутствует"),
            "stars": data.get("stargazers_count", 0),
            "language": data.get("language", "Не указан")
        }
    except Exception as e:
        return {"error": f"Ошибка API: {str(e)}"}
```

```python
def analyze_repo(state):
    if "error" in state:
        return state

    description = state.get("description", "")
    prompt = f"""Объясни простыми словами что делает этот GitHub проект:

{description}

Ответ должен быть кратким (2-3 предложения) и понятным для начинающих разработчиков."""

    try:
        response = llm.invoke(prompt)
        state["analysis"] = response.content
        return state
    except Exception as e:
        state["error"] = f"Ошибка анализа: {str(e)}"
        return state
```

```python
def format_response(state):
    if "error" in state:
        return {"result": f"Ошибка: {state['error']}"}

    result = f"""
Репозиторий: {state.get('repo')}

Описание: {state.get("description")}

Язык: {state.get("language")}

Stars: {state.get("stars")}

Анализ проекта:
{state.get("analysis")}
"""
    return {"result": result}
```

```python
builder = StateGraph(dict)

builder.add_node("api", github_api)
builder.add_node("analyze", analyze_repo)
builder.add_node("format", format_response)

builder.set_entry_point("api")
builder.add_edge("api", "analyze")
builder.add_edge("analyze", "format")

graph = builder.compile()
```

```python
MODEL_NAME = "Qwen/Qwen2-VL-2B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
)

print("Загрузка модели")
model = Qwen2VLForConditionalGeneration.from_pretrained(
    MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
)

processor = AutoProcessor.from_pretrained(
    MODEL_NAME,
    min_pixels=256 * 28 * 28,
    max_pixels=2048 * 28 * 28,
)
print("Модель загружена!")
```

```python
def preprocess_image(path):

    img = cv2.imread(path)
    if img is None:
        raise ValueError(f"Не удалось загрузить: {path}")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    mean_brightness = np.mean(gray)
    std_brightness = np.std(gray)

    result = img.copy()

    # Коррекция контраста, но если его мало
    if std_brightness < 30:
        lab = cv2.cvtColor(result, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        l = clahe.apply(l)
        lab = cv2.merge([l, a, b])
        result = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

    # Осветление тёмных изображений
    if mean_brightness < 80:
        hsv = cv2.cvtColor(result, cv2.COLOR_BGR2HSV)
        h, s, v = cv2.split(hsv)
        v = cv2.add(v, 80)
        hsv = cv2.merge([h, s, v])
        result = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    return Image.fromarray(result_rgb)
```

```python
def run_vlm_ocr(pil_image):
    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "image": pil_image},
                {
                    "type": "text",
                    "text": (
                        "Внимательно прочитай весь рукописный текст на изображении. "
                        "Выведи только распознанный текст, без комментариев, "
                        "без описания изображения, без пояснений. "
                        "Сохрани оригинальную пунктуацию и переносы строк."

                    ),
                },
            ],
        }
    ]

    text_prompt = processor.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )
    image_inputs, video_inputs = process_vision_info(messages)

    inputs = processor(
        text=[text_prompt],
        images=image_inputs,
        videos=video_inputs,
        padding=True,
        return_tensors="pt",
    ).to(model.device)

    with torch.no_grad():
        generated_ids = model.generate(
            **inputs,
            max_new_tokens=1024,
            do_sample=False,
            repetition_penalty=1.1,
        )

    generated_ids_trimmed = [
        out_ids[len(in_ids):]
        for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]

    output_text = processor.batch_decode(
        generated_ids_trimmed,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=False,
    )[0]

    return output_text.strip()
```

```python
print("\n" + "=" * 60)
print("РАСПОЗНАВАНИЕ ТЕКСТА")
print("=" * 60)

hyps = []

for i, img_path in enumerate(image_files):
    print(f"\n[{i+1}/{len(image_files)}] Обработка: {os.path.basename(img_path)}")
    print("-" * 50)

    try:
        pil_image = preprocess_image(img_path)
        print(f"  Размер изображения: {pil_image.size}")
        text = run_vlm_ocr(pil_image)
        hyps.append(text)
        print(f"Распознанный текст:")
        print(text)

    except Exception as e:
        print(f"ОШИБКА: {str(e)}")
        hyps.append("")

    print("-" * 50)
```
